```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = {
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': [5, 4, 3, 2, 1],
    'D': [8, 6, 2, 3, 7]
}
df_clean = pd.DataFrame(data)


sns.pairplot(df_clean)
plt.show()
plt.figure(figsize=(8, 6))
sns.heatmap(df_clean.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(data=df_clean)
plt.title('Boxplot')
plt.show()
```
